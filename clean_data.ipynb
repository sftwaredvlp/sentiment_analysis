```python
import pandas as pd
import numpy as np 
from nltk.corpus import stopwords
```

```python
tweets=pd.read_csv("scraped_tweets.csv")
tweets.head()
```

```python
tweets.drop(["date","likes","comments","retweets","twitter_link","username"], axis=1, inplace=True)
tweets.head()
```

```python
import re

# Define a function to remove emojis and punctuation
def remove_emoji_and_punctuation(text):
    # Remove emojis
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F700-\U0001F77F"  # alchemical symbols
                               u"\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
                               u"\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
                               u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
                               u"\U0001FA00-\U0001FA6F"  # Chess Symbols
                               u"\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
                               u"\U00002702-\U000027B0"  # Miscellaneous Symbols
                               u"\U000024C2-\U0001F251" 
                               "]+", flags=re.UNICODE)
    text = emoji_pattern.sub(r'', text)

    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    return text

# Apply the remove_emoji_and_punctuation function to each cell in the 'text' column
tweets['text'] = tweets['text'].apply(remove_emoji_and_punctuation)

tweets.head()
```

```python
import re
import nltk
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.corpus import words

nltk.download('punkt')
nltk.download('words')

english_words = set(words.words())

# Define a function to clean and filter words in a given text
def clean_and_filter_words(text, english_words):
    
    words = word_tokenize(text)
    cleaned_words = []

    for word in words:
        # Remove digits, non-alphabetic characters, and non-English characters
        cleaned_word = re.sub(r'[^a-zA-Z]', '', word)

        # Check if the cleaned word is an English word and has a length greater than 2
        if cleaned_word.lower() in english_words and len(cleaned_word) > 2:
            # Append the cleaned word to the list (strip() removes leading and trailing spaces)
            cleaned_words.append(cleaned_word.lower().strip())

    return ' '.join(cleaned_words)

# Apply the clean_and_filter_words function to each cell in the 'text' column
tweets['text'] = [clean_and_filter_words(text, english_words) for text in tweets['text']]

```

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
# Remove stopwords from the 'text' column in the Text column
tweets['text'] = tweets['text'].apply(lambda x: ' '.join([word for word in x.split() if word.lower() not in stop_words]))
```
