    Handling Missing Values:
        * Fill missing values in the "text" column with an empty string to ensure compatibility with the CountVectorizer.
    Entered Text for Prediction:
        * Define the entered text for which you want to find similar hashtags.
    Sentiment Analysis for Entered Text:
        * Use TextBlob to calculate the sentiment score for the entered text.
    CountVectorizer:
        * Create a CountVectorizer instance to convert the text data into a matrix of token counts.
    Cosine Similarity Calculation:
        * Transform the text data using CountVectorizer and calculate cosine similarity between the entered text and each row in the text matrix.
    Combining Similarity and Sentiment Scores:
        * Combine the similarity scores with the sentiment score of the entered text for ranking.
    Top Three Similar Texts:
        * Get the indices of the top three most similar texts.
    Retrieve Hashtags and Sentiment Scores:
        * Retrieve corresponding hashtags and sentiment scores for the top three similar texts.
    Remove Duplicate Hashtags:
        * Remove duplicates from recommended hashtags.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from textblob import TextBlob
import pandas as pd

# Fill missing values in the "text" column with an empty string
final_data["text"].fillna("", inplace=True)

# Entered text for prediction
entered_text = "Life is either or complicated"

# Calculate sentiment score for the entered text
entered_text_sentiment = TextBlob(entered_text).sentiment.polarity
entered_text_subjectivity = TextBlob(entered_text).sentiment.subjectivity
entered_text_sentiment_score = entered_text_sentiment * entered_text_subjectivity

# Create a CountVectorizer instance
vectorizer = CountVectorizer()

# Fit and transform the text data
text_matrix = vectorizer.fit_transform(final_data["text"])

# Calculate cosine similarity between the entered text and each row in the text matrix
entered_text_vectorized = vectorizer.transform([entered_text])
similarity_scores = cosine_similarity(entered_text_vectorized, text_matrix).flatten()

# Combine similarity scores with sentiment scores for ranking
combined_scores = similarity_scores + entered_text_sentiment_score * final_data["sentiment_score"]

# Get the indices of the top three most similar texts
top_indices = combined_scores.argsort()[-3:][::-1]

# Get the corresponding hashtags and sentiment scores for the top three similar texts
recommended_hashtags = final_data.loc[top_indices, "hashtag"].tolist()
recommended_sentiment_scores = final_data.loc[top_indices, "sentiment_score"].tolist()

# Remove duplicates from recommended hashtags
unique_recommended_hashtags = list(set(recommended_hashtags))

print("Entered Text:", entered_text)
print("Entered Text Sentiment Score:", entered_text_sentiment_score)
print("Recommended Hashtags:", unique_recommended_hashtags)
print("Recommended Sentiment Scores:", recommended_sentiment_scores)
```
