```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data_sentiment.csv")
# Hashtag frequency 
plt.figure(figsize=(20, 10))
df['hashtag'].value_counts().plot(kind='bar', color='orchid')
plt.title('Hashtag Histogram')
plt.xlabel('Hashtag')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
from wordcloud import WordCloud

# combining the all hashtags where in hashtag column
all_hashtags = ' '.join(df['hashtag'])

# Create WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='black', 
                      colormap='viridis', collocations=False).generate(all_hashtags)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud')
plt.show()

```

```python
# Sentiment analysis frequency histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='sentiment', hue='sentiment', multiple='stack', palette='viridis', shrink=0.8)
plt.title('Sentiment Analysis Result')
plt.xlabel('Sentiment')
plt.ylabel('Frequency')
plt.show()
```

```python
# Hashtag's frequency acoording to sentiment degrees
plt.figure(figsize=(50, 20))
sns.barplot(data=df, x='hashtag', y='sentiment_score', palette='viridis')
plt.title('Sentiment Degrees with Hashtags')
plt.xlabel('Hashtag')
plt.ylabel('Sentiment Degree')
plt.xticks(rotation=90)
plt.show()
```
